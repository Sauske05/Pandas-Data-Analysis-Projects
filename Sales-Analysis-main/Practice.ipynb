```python
import pandas as pd
import os 
practice_data = pd.DataFrame()

files = [file for file in os.listdir('./Sales_Data/')]
print(files)
```

```python
for file in files :
    df = pd.read_csv('./Sales_Data/' + file)
    practice_data = pd.concat([practice_data, df])
    
practice_data.to_csv('practice.csv', index= False)
```

```python
df = pd.read_csv('practice.csv')
df.head()
```

```python
df = df.dropna(how='any')
df
```

```python
def get_month(month) :
    return month.split("/")[0]

df['Month'] = df['Order Date'].apply(get_month)
df
```

```python
df.drop(df[(df['Order Date'].str.contains('O'))].index, inplace=True)
```

```python
df
```

```python
def get_month(month) :
    return month.split("/")[0]

df['Month'] = df['Order Date'].apply(get_month)
df
```

df['Month'] = df['Month'].astype(int32)

```python
df['Month'] = df['Month'].astype(int)
df
```

```python
df['Quantity Ordered'] = df['Quantity Ordered'].astype(int)
df['Price Each'] = df['Price Each'].astype(float)

df['Total Price'] = df['Quantity Ordered'] * df['Price Each']
```

```python
df
```

```python
highest_sales_bymonth = df.groupby(['Month']).sum()
highest_sales_bymonth
```

```python
import matplotlib.pyplot as plt
months = [month for month, df in df.groupby('Month')]
plt.bar(months, highest_sales_bymonth['Total Price'])
plt.xticks(months)
plt.ylabel('Sales')
plt.xlabel('Month')
plt.show()
```

```python
def get_city(city) :
    return city.split(",")[1]
def get_state(state) :
    return state.split(",")[2].split(" ")[1]

df['City'] = df['Purchase Address'].apply(lambda x : get_city(x)+"(" + get_state(x) + ")")
df
```

```python
city_sales = df.groupby(['City']).sum()
city_sales
```

```python
cities = [city for city, df in df.groupby('City')]
plt.bar(cities, city_sales['Total Price'])
plt.xticks(cities, rotation= 'vertical')
plt.ylabel('Sales')
plt.xlabel('City')
plt.show()
```
