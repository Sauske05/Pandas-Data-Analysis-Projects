```python
df['Month'] = df['film_date'].dt.month
df['Month_Name'] = df['film_date'].dt.month_name()
df
```

```python
df['count'] = 1
df
```

```python
month_analysis = df.groupby(['Month_Name']).sum()
month_analysis.sort_values(by=['count'], ascending=False)
```

```python
plt.figure(figsize=(15,6))
months = [month for month, df in df.groupby('Month_Name')]
print(months)
sns.barplot(x=months, y='count', data=month_analysis)
plt.show()
```
