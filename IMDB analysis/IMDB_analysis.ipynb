```python
import pandas as pd
import numpy as np
df = pd.read_csv('IMDB Top 250 Movies.csv')
df
```

```python
df.columns
```

```python
df = df[['rank', 'name', 'year', 'rating', 'genre',
#                 'certificate',
                'run_time',
#        'tagline',
                'budget', 'box_office',
#                 'casts', 'directors', 'writers'
               ]]
```

```python
df
```

```python
df = df.rename(columns= {'name' : 'Name', 'rank' : 'Rank', 'year' : 'Year', 'rating' : 'Rating', 'genre' : 'Genre', 'budget' : 'Budget', 'run_time' : 'Run_time', 'box_office' : 'Box_office'})
```

```python
df
```

```python
 df.isna().sum()
```

```python
df = df.replace('Not Available', np.nan).dropna()
df
```

```python
df.dtypes
```

```python
df['Budget'] = df['Budget'].str.replace(r'[^0-9]', '')
df
```

```python
df['Budget'] = pd.to_numeric(df['Budget'])
df.dtypes
```

```python
df['Box_office'] = pd.to_numeric(df['Box_office'])
```

```python
df['Box_office'] = df['Box_office'].str.replace(r'[^0-9]', '')
```

```python
df
```

```python
df = df.sort_values(by=['Box_office'], ascending=False)
df['Profit'] = df['Box_office'] - df['Budget']
df = df.reset_index(drop= True)
df
```
