```python
import pandas as pd
import os
all_data = pd.DataFrame()

files = [file for file in os.listdir('./Sales_Data')]
type(files)
```

```python
print(files)
```

```python
for index in files:
    df = pd.read_csv('Sales_Data/' + index) #'Sales_Data/Sales_April_2019.csv'
    all_data = pd.concat([all_data, df])
    
all_data.to_csv('all_data.csv',index=False)
all_data.tail()
```
