```python
new_df = pd.read_csv('filtered_data.csv')
```

```python
new_df
```

```python
new_df['Quantity Ordered'] = new_df['Quantity Ordered'].astype(int)
```

```python
new_df['Price Each'] = new_df['Price Each'].astype(float)
```

```python
new_df['sales'] = new_df["Quantity Ordered"] * new_df['Price Each']
```

```python
new_df['sales'] = new_df['sales'].astype(float)
```

```python
def get_city(address) :
    return address.split(',')[1]

def get_state(address) :
    return address.split(',')[2].split(' ')[1]

new_df['City'] = new_df['Purchase Address'].apply(lambda x: get_city(x) + "(" + get_state(x) + ")")
new_df.head()
```

```python
results = new_df.groupby('City').sum()
results
```

```python
import matplotlib.pyplot as plt
cities = [city for city, df in new_df.groupby('City')]
print(cities)
plt.bar(cities, results['sales'])
plt.xticks(cities, rotation = 'vertical')
plt.ylabel('Sales')
plt.xlabel('City')
plt.show()
```
